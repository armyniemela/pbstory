```python
# want to see a list of unique brand names
brand = df['Brand'].str.split(', ')
brand = [brand for sublist in brand for brand in sublist]
unique_brands = set(brand)
unique_brands
```

```python
df["Brand"] = (df["Brand"]
    .str.replace("JIFF", "JIF")
    .str.replace("PeanutButterCo", "PeanutButter&Co" )
    .str.replace("Justines", "Justins'" )
    .str.replace("CreamNut ", "CreamNut" )
    .str.replace("EssentialseverydayPB", "EssentialseverydayPeanutButter" )
    .str.replace("CBsNuts", "CB'sNuts")
)
df
```

```python
df["Brand"] = (df["Brand"]
        .str.replace("EssentialseverydayPeanutButter", "EssentialsEverydayPeanutButter")
        .str.replace("UrbanMeadow", "Urbanmeadow")
        .str.replace("OnceAgain", "Once Again")
        .str.replace("Brads", "Brad's")
        .str.replace("CreamNut", "Creamnut")
        .str.replace("SmartBalance", "Smartbalance")
        .str.replace(" PB2", "PB2")
        .str.replace("Wow Butter", "WowButter")
        .str.replace("Reeses", "Reese's")
        .str.replace("Ricards100%Peanuts", "Ricards 100% Peanuts")        
)
```

```python
import re
```

```python
def split_brand(brands):
    return re.sub(r'(?<=[a-z])(?=[A-Z])', ' ', brands)

df['Brand'] = df['Brand'].apply(split_brand)
df[['Brand']]
```

```python
df.head()
```

```python
# chatGPT helped me with this: make a list from the Brand-column and count the unique brands.

brands = df['Brand'].str.split(', ')
all_brands = [brand 
    for sublist in brands 
    for brand in sublist
]
brands = set(all_brands)
brand_count = len(brands)
print(f"Unique brands count: {brand_count}")
```

```python
# getting the brands sold in each store (mean)

df['Unique_Brands'] = df['Brand'].str.split(', ').apply(lambda x: len(set(x)))
mean_brands = df['Unique_Brands'].mean()
print(f"Unique brands per store: {mean_brands}")
```
