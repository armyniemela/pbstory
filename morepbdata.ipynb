```python
# Split the 'Brand' column and explode the DataFrame
peanut_df['Brand'] = peanut_df['Brand'].str.split(', ')
peanut_df = peanut_df.explode('Brand')
pivot_df = peanut_df.pivot_table(index='Brand', columns='Store', aggfunc=lambda x: 1.0, fill_value=0)
pivot_df = pivot_df.reset_index()
pivot_df
```

```python
pivot_df = pivot_df[pivot_df['Brand'] != '']
pivot_df = pivot_df.drop_duplicates(subset='Brand', keep='first')
#pivot_df
```

```python
pivot_df = pivot_df[pivot_df['Brand'] != 'Richards100%Peanuts']
pivot_df
```

```python
pivot_df.dtypes
```

```python
pivot_df.head()
```

```python
pivot_df.iloc[:, 1:] = pivot_df.iloc[:, 1:].astype(int)
pivot_df = pivot_df.convert_dtypes()
pivot_df
```

```python
brand_store_counts = pivot_df.iloc[:, 1:].sum(axis=1)
pivot_df['Store Count'] = brand_store_counts
pivot_df
```

```python
sorted_pivot_df = pivot_df.sort_values(by='Store Count', ascending=False)
sorted_pivot_df
```

```python
total_stores = len(pivot_df.columns) - 2
skippy_row = pivot_df[pivot_df['Brand'] == 'Skippy']
skippy_count = skippy_row['Store Count'].values[0]
skippy_percentage = (skippy_count / total_stores) * 100
skippy_percentage
```
